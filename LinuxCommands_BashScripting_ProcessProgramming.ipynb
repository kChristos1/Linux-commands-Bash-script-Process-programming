**Ερώτημα 1.7: Φτιάξτε με command chaining σε μια γραμμή, ένα νέο αρχείο με όνομα `motes_10_20.txt` που θα περιέχει δύο στήλες, ως εξής:**
* **η πρώτη στήλη θα έχει τις μετρήσεις θερμοκρασίας του sensor 10 από το `temperature.txt`**
* **η δεύτερη στήλη θα έχει τις μετρήσεις υγρασίας του sensor 10 από το `humidity.txt`**
* **η τρίτη στήλη θα έχει τις μετρήσεις θερμοκρασίας του sensor 20 από το `temperature.txt`**
* **η τέταρτη στήλη θα έχει τις μετρήσεις υγρασίας του sensor 20 από το `humidity.txt`**
* **οι στήλες στο `motes_10_20.txt` θα χωρίζονται μεταξύ τους με το χαρακτήρα `:`**
* **οι γραμμές στο `motes_10_20.txt` θα πρέπει να είναι ταξινομημένες με βάση τα νούμερα στη 2η στήλη λαμβάνοντας υπόψιν τα 2 δεκαδικά. Για παράδειγμα για τιμές 21.235 και 21.237 θα λάβουμε υπόψη μόνο το 21.23**

````python
!paste <(cut -d ' ' -f 10 /content/drive/MyDrive/files_proj2/temperature.txt) <(cut -d ' ' -f 10 /content/drive/MyDrive/files_proj2/humidity.txt) <(cut -d ' ' -f 20 /content/drive/MyDrive/files_proj2/temperature.txt) <(cut -d ' ' -f 20 /content/drive/MyDrive/files_proj2/humidity.txt) | tr '\t' ':' | sort -t':' -k2,2n > motes_10_20.txt
````

Σχολιάστε τι κάνει η παραπάνω εντολή ή chained εντολές που δώσατε:

Oι commands:

cut -d ' ' - f 10 temperature.txt

cut -d ' ' - f 10 humidity.txt

cut -d ' ' - f 20 temperature.txt

cut -d ' ' - f 20 humidity.txt

Χρησιμοποιούνται για να κάνουν extract τις θερμοκρασίες των sensor 10 και 20 αλλά και τις μετρήσεις υγρασίας των sensor 10 και 20, από τα αντίστοιχα αρχεία temperature.txt και humidity.txt. Το -d θέτει το delimeter ως whitespace character.

Η command paste <(...) <(...) <(...) <(...) , κάνει paste το output των τεσσάρων cut commands. Δηλαδή τα αποτελέσματα των cut, θα είναι "δίπλα-δίπλα".

Με χρήση pipe , το αποτέλεσμα της paste περνάει μέσα από το pipe.

Μέσω του pipe το αποτέλεσμα αυτό περνάει ως είσοδος στην εντολή tr (βλ. παρακάτω).


Με τη χρήση της tr '\t' αντικαθιστώ τα tab characters με το επιθυμητό delimeter ":".

Όπως ζητείται από την εκφώνηση κάνω sort τα αποτελέσματα βάσει των τιμών που υπάρχουν στη δεύτερη στήλη.Η ταξινόμιση γίνεται αριθμητικά.

Η commad sort -t':' -k2,2n ταξινομεί τις γραμμές βάσει των τιμών της δεύτερης στήλης (-k2) και ως delimeter που διαχωρίζει τις τιμές ,θεωρείται το ":" (-t':').
Το -k2,2 σημαίνει ότι κάνω sorting βάσει του δεύτερου πεδίου (στήλης μόνο.
To n σημαίνει ότι κάνω numerical sorting.

````python
%%file aggregative_news_analyzer.sh
#!/bin/bash

# Returns a subset of doc_ids for a specific tid
doc_t() {
    target_tid="$1"
    term_list="$2"
    subset=()

    for item in "${term_list[@]}"; do
        doc_id="${item%% *}"  # Extract the doc_id before the first space
        tids="${item#*}"     # Extract the part after the first ' ' to get the tids
        IFS=' ' read -ra tid_array <<< "$tids"

        for tid in "${tid_array[@]}"; do
            # Extract the actual tid by removing anything after ":"
            actual_tid="${tid%%:*}"

            if [ "$actual_tid" == "$target_tid" ]; then
                subset+=("$doc_id")
                break
            fi
        done
    done
    echo "${subset[@]}"
}

# Returns a subset of doc_ids for a specific category
doc_c() {
    target_category="$1"
    category_list="$2"
    subset=()
    for item in "${category_list[@]}"; do
        IFS=' ' read -ra category_docid <<< "$item"
        categories="${category_docid[0]}"
        docid="${category_docid[1]}"
        IFS=' ' read -ra category_array <<< "$categories"
        for category in "${category_array[@]}"; do
            if [ "$category" == "$target_category" ]; then
                subset+=("$docid")
                break
            fi
        done
    done
    echo "${subset[@]}"
}



# Calculate the intersection of two lists
list_intersection() {
    local list1=($1)
    local list2=($2)
    local intersection=()
    for item in "${list1[@]}"; do
        if [[ " ${list2[@]} " =~ " $item " ]]; then
            intersection+=($item)
        fi
    done
    echo "${intersection[@]}"
}

# Calculate the union of two lists
list_union() {
    local list1=($1)
    local list2=($2)
    local union=()
    union+=("${list1[@]}")
    for item in "${list2[@]}"; do
        if [[ ! " ${list1[@]} " =~ " $item " ]]; then
            union+=($item)
        fi
    done
    echo "${union[@]}"
}

# Calculate the cardinality of the given list
cardinality() {
    local my_list=($1)
    echo "${#my_list[@]}"
}

# Calculate the Jaccard Index
jaccard_index() {
    local list1=($1)
    local list2=($2)
    local intersection=($(list_intersection "${list1[*]}" "${list2[*]}"))
    local union=($(list_union "${list1[*]}" "${list2[*]}"))
    local intersection_cardinality=$(cardinality "${intersection[*]}")
    local union_cardinality=$(cardinality "${union[*]}")
    if (( union_cardinality != 0 )); then
        index=$(bc -l <<< "$intersection_cardinality / $union_cardinality")
    else
        index=0
    fi
    if (( $(bc -l <<< "$index < 0 || $index > 1") )); then
        result=-1
    else
        result=$index
    fi
    echo "$result"
}

# Find tid in stem_tid_map for a specific stem
find_tid() {
    target_stem="$1"
    stem_tid_map="$2"
    term_id=""
    for item in "${stem_tid_map[@]}"; do
        IFS=' ' read -ra stem_tid <<< "$item"
        stem="${stem_tid[0]}"
        tid="${stem_tid[1]}"
        if [ "$stem" == "$target_stem" ]; then
            term_id="$tid"
            break
        fi
    done
    echo "$term_id"
}

# Calculate jaccard index for stem - category
calculate_jaccard_index_for_stem_category() {
    stem="$1"
    category="$2"
    category_list="$3"
    term_list="$4"
    stem_tid_map="$5"
    term_id=$(find_tid "$stem" "${stem_tid_map[*]}")
    term_documents=$(doc_t "$term_id" "${term_list[*]}")
    category_documents=$(doc_c "$category" "${category_list[*]}")
    jaccard_score=$(jaccard_index "$term_documents[*]" "$category_documents[*]")
    echo "$jaccard_score"
}

# Show stems for a spesific doc_id
show_stems_for_doc_id() {
    doc_id=$1

    tids=( $(grep "^$doc_id " "$term_file" | awk '{
        for (i = 2; i <= NF; i++) {
            split($i, a, ":");
            print a[1];
        }
    }') )

    if [ ${#tids[@]} -eq 0 ]; then
        echo "There is no tid for doc_id $doc_id."
        return
    fi


    for tid in "${tids[@]}"; do
        echo "Results for tid $tid:"
        grep_result=$(grep " $tid " "$stem_tid_file" | cut -d ' ' -f 1 | uniq)
        if [ -z "$grep_result" ]; then
            echo "There is no stem tid $tid."
        else
            echo "$grep_result"
        fi
    done
}

# Show categories for a spesific doc_id
show_categories_for_doc_id() {
    did=$1
    result=$(grep " $did " "$category_file" | cut -d ' ' -f 1)
    if [ -z "$result" ]; then
        echo "There is no category for the doc_id $did."
    else
        echo "$result"
    fi
}

# Count terms for a spesific doc_id
count_terms_of_doc() {
    doc_id=$1
    result=$(grep "^$doc_id " "$term_file" | awk '{
        for (i = 2; i <= NF; i++) {
            split($i, a, ":");
            print a[1];
        }
    }')

    if [ -z "$result" ]; then
        echo "There is no tid for doc_id $doc_id."
    else
        echo "There are $(echo "$result" | wc -l) tids for the doc_id $doc_id"
    fi
}


# Count categories for a spesific doc_id
count_categories_of_doc() {
    doc_id=$1
    result=$(grep " $doc_id " "$category_file" | cut -d ' ' -f 1)

    if [ -z "$result" ]; then
        echo "There is no category for the doc_id $doc_id."
    else
        echo "There are $(echo "$result" | wc -l) categories for the doc_id $doc_id."
    fi
}

# This function shows stems or categories if user give input = -t or -c
handle_option1() {
    option=$1
    did=$2

    if [ "$option" == "-t" ]; then
        echo "Option '-t' selected."
        show_stems_for_doc_id "$did"
    elif [ "$option" == "-c" ]; then
        echo "Option '-c' selected."
        show_categories_for_doc_id "$did"
    else
        echo "Invalid option."
    fi
}

# This function counts tids or categories if user gives input = -t or -c
handle_option2() {
    option=$1
    did=$2

    if [ "$option" == "-t" ]; then
        echo "Option '-t' selected."
        count_terms_of_doc "$did"
    elif [ "$option" == "-c" ]; then
        echo "Option '-c' selected."
        count_categories_of_doc "$did"
    else
        echo "Invalid option."
    fi
}

menu()
{
    echo "-------------------------------- Aggregative News Analyzer Menu --------------------------------"
    echo "                                                                                                "
    echo "---------Commands-------------------------------Details-----------------------------------------"
    echo "-     $ <stem> <category>        Show me the Jaccard Index of the specific (stem,              -"
    echo "-                                category) pair                                                -"
    echo "------------------------------------------------------------------------------------------------"
    echo "-      P <did> -c|-t             If the “-t” switch is used in the command, then show me all   -"
    echo "-                                the stems contained in the document with code did. If the     -"
    echo "-                                “-c” switch is used in the command, then show me all the      -"
    echo "-                                categories assigned to the document with code did.            -"
    echo "------------------------------------------------------------------------------------------------"
    echo "-      C <did> -c|-t             If the “-t” switch is used in the command, count and show me  -"
    echo "-                                on the screen how many (count) unique terms the document      -"
    echo "-                                with code did contains. If the “-c” switch is used in the     -"
    echo "-                                command, count and show me on the screen how many (count)     -"
    echo "-                                categories are assigned to the document with code did.        -"
    echo "------------------------------------------------------------------------------------------------"
    echo "-       X                        Exit                                                          -"
    echo "------------------------------------------------------------------------------------------------"

    PS3="Enter your command (press 1 for $, 2 for P , 3 for C or 4 for X ):"
    select command in "$" "P" "C" "X"; do
        case $command in
            "$" )
                echo "Option '$' selected."
                read -p "Please enter the stem: " stem
                read -p "Please enter the category: " category
                jaccard_score=$(calculate_jaccard_index_for_stem_category "$stem" "$category" )
                echo "Jaccard Index for stem '$stem' and category '$category': $jaccard_score"
                echo "Action completed!"
                menu
                ;;
            "P" )
                echo "Option 'P' selected."
                read -p "Please enter the document_id: " did
                read -p "Please enter '-t' or '-c': " option
                handle_option1 "$option" "$did"
                echo "Action completed!"
                menu
                ;;
            "C" )
                echo "Option 'C' selected."
                read -p "Please enter the document_id: " did
                read -p "Please enter '-t' or '-c': " option
                handle_option2 "$option" "$did"
                echo "Action completed!"
                menu
                ;;
            "X" )
                echo "Thanks for using our program!"
                exit
                ;;
            * )
                echo "Please enter a valid input."
                ;;
        esac
    done
}


main() {

    # Define the paths for category, term, and stem-tid  files
    read -p "Enter the path for the category file: " category_file
    read -p "Enter the path for the term file: " term_file
    read -p "Enter the path for the stem-tid file: " stem_tid_file

    # Read the contents of the files into arrays
    readarray  category_list < "$category_file"
    readarray  term_list < "$term_file"
    readarray  stem_tid_map < "$stem_tid_file"

    menu
}

main
````

Γράψτε ένα πρόγραμμα Bash Script που διαβάζει ένα-ένα κάθε document (δλδ μια-μια κάθε γραμμή) του `lyrl2004_vectors_train.dat.txt`, όπου το document id είναι η πρώτη τιμή σε κάθε γραμμή, και για κάθε γραμμή που διαβάζει:
* Αντιστοιχεί τα term ids που περιέχει με τα stems στο `stem.termid.idf.map.txt`

* Γράφει σε ένα αρχείο `sentimentpernewsitem.txt`, για κάθε document που επεξεργάζεται, το επικρατές sentiment. Π.χ. για τo document στη γραμμή 1 του `lyrl2004_vectors_train.dat.txt`, στο `sentimentpertweet.txt` θα γραφτεί:
  * `Document 2286 has a positive sentiment` αν ο αριθμός των positive stem στο document είναι μεγαλύτερος από τον αριθμό των negative
  * `Document 2286 has a neutral sentiment` αν ο αριθμός των negative stem στο document είναι ίσος με τον αριθμό των positive
  * `Document 2286 has a negative sentiment` αν δεν ισχύει κάτι από τα παραπάνω

## ⏩ Bash Script Κώδικας του News Sentiment Analyser

````python
%%file news_sentiment_analyser.sh
#!/bin/bash
input="/content/drive/MyDrive/ERGALEIA/files_proj2/lyrl2004_vectors_train10.dat.txt"
pos="/content/drive/MyDrive/ERGALEIA/files_proj2/positive_stems.txt"
neg="/content/drive/MyDrive/ERGALEIA/files_proj2/negative_stems.txt"

output="/content/sentimentpernewsitem.txt"

sentiment_counter="0"

pos_arr=() #empty array to store positive stems
while IFS= read -r lin; do
  pos_arr+=("$lin")
done < "$pos"


neg_arr=()
while IFS= read -r lin; do
  neg_arr+=("$lin")
done < "$neg"

#read the data from the input file and store it in all_lines array.
readarray  all_lines < "$input"


for item in "${all_lines[@]}"; do
    #in this array we hold the tids of the specific document with $docid
    tids_for_doc_for_line=()
    stems_for_doc_for_line=()

    #"break" the line to pieces. docid holds the docid and tids holds the rest..
    IFS=' ' read -ra docid_tid <<< "$item"
    #Extracting docid...
    docid="${docid_tid[0]}"
    #Extracting tids...
    tids="${docid_tid[@]:1}"
    #Crop and keep only the tids, removing the garbage values.
    IFS=' ' read -ra tids_for_doc_for_line <<< "$(echo "$tids" | sed 's/:[^ ]*//g')"


    for term_id in "${tids_for_doc_for_line[@]}"; do
      #i-th termid
      #with the following instruction i extract the stem that matches the term_id:
      stem=$(grep " $term_id " /content/drive/MyDrive/ERGALEIA/files_proj2/stem.termid.idf.map.txt | cut -d' ' -f1)

      stems_for_doc_for_line+=("$stem")
    done

    for general_stem in "${stems_for_doc_for_line[@]}"; do
      #echo "general stem = $general_stem"
      for positive_stem in "${pos_arr[@]}"; do
        #echo "positive stem = $positive_stem"
        if [ "${positive_stem}" == "${general_stem}" ]; then
            let "sentiment_counter+=1"
            #echo "ctr: $sentiment_counter"
        fi
      done
      for negative_stem in "${neg_arr[@]}"; do
        #echo "negative_stem stem = $negative_stem"
        if [ "${negative_stem}" == "${general_stem}" ]; then
            let "sentiment_counter-=1"
            #echo "ctr : $sentiment_counter"
        fi
      done
    done
    if [[ $sentiment_counter -gt "0" ]]; then
      echo "Document ${docid} has a positive sentiment" >> "$output"
    elif [[ $sentiment_counter -lt "0" ]]; then
      echo "Document ${docid} has a negative sentiment" >> "$output"
    elif [[ $sentiment_counter -eq "0" ]]; then
      echo "Document ${docid} has a neutral sentiment" >> "$output"
    fi
    let "sentiment_counter=0"
done
````

## ⏩ Κώδικας Mixed Piping

Γράψτε στο παρακάτω code cell όλο τον κώδικα του three-way pipe σας.

````python
%%file mixed_piping.py
import os
import sys
import time

fifoname = 'pipefifo'

MESSAGES = (
    "I love PLH211          ",
    "The professor is boring",
    "but the curriculum is  ",
    "interesting & useful   "
)

def source_process(pipe_write):
    try:
        for message in MESSAGES:
            os.write(pipe_write, message.encode())
            time.sleep(1)  # Simulate some processing time
        os.close(pipe_write)
    except Exception as e:
        print(f"Error in source process: {e}")
        os._exit(1)

def transformer_process(pipe_read, fifo_write):
    try:
        with os.fdopen(pipe_read, 'r') as pipe_reader:
            for message in MESSAGES:
                data = pipe_reader.readline()
                tokens = data.split()
                transformed_message = "\n".join(tokens)  #seperate each word with a newline character.
                fifo_write.write(transformed_message + "\n")
                fifo_write.flush()

        fifo_write.close()
    except Exception as e:
        print(f"Error in transformer process: {e}")
        os._exit(1)


def output_process(fifo_read):
    try:
        while True:
            data = fifo_read.readline()
            if not data:
                break
            print(data, end="")
            sys.stdout.flush()
        fifo_read.close()
    except Exception as e:
        print(f"Error in output process: {e}")
        os._exit(1)

def main():
    try:
        # Remove existing FIFO if it exists
        if os.path.exists(fifoname):
            os.unlink(fifoname)

        # Create FIFO (named pipe)
        os.mkfifo(fifoname)

        pipe_read, pipe_write = os.pipe()

        source_pid = os.fork()
        if source_pid == 0:
            os.close(pipe_read)
            source_process(pipe_write)
            sys.exit(0)

        transformer_pid = os.fork()
        if transformer_pid == 0:
            os.close(pipe_write)
            fifo_write = open(fifoname, "w")
            transformer_process(pipe_read, fifo_write)
            sys.exit(0)

        output_pid = os.fork()
        if output_pid == 0:
            os.close(pipe_read)
            os.close(pipe_write)
            fifo_read = open(fifoname, "r")
            output_process(fifo_read)
            sys.exit(0)

        # Close unnecessary file descriptors
        os.close(pipe_read)
        os.close(pipe_write)

        # Wait for processes to finish
        os.waitpid(source_pid, 0)
        os.waitpid(transformer_pid, 0)
        os.waitpid(output_pid, 0)

        # Remove FIFO
        os.unlink(fifoname)

    except Exception as e:
        print(f"Error in main: {e}")

if __name__ == "__main__":
    sys.exit(main())
````
